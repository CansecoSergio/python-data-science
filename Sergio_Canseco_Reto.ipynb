Importar frameworks

```python
import pandas as pd
```

**Recuperar la información del Archivo employee_attrition.csv**

```python
main_df = pd.read_csv('/content/drive/MyDrive/Archivos Collab/Employee Attrition.csv')

main_df
```

**El tiempo de servicio promedio de los trabajadores**

```python
#main_df.info()
servicio_promedio = main_df['Time_of_service'].mean()
servicio_promedio_df = pd.DataFrame({'Tiempo_servicio_promedio': [servicio_promedio]})

servicio_promedio_df
```

**El porcentaje de hombres y mujeres en el archivo**

```python
#main_df.info()
porcentaje_df = main_df[['Gender']].groupby('Gender').value_counts().reset_index(name='Total_gender')
total_empleados = porcentaje_df['Total_gender'].sum()

porcentaje_df['Procentaje'] = (porcentaje_df['Total_gender'] * 100) / total_empleados
porcentaje_df = porcentaje_df[['Gender', 'Procentaje']]

porcentaje_df
```

**El tiempo máximo sin promoción de las personas que llevan más de 15 años de servicio**

```python
#main_df.info()
tiempo_spromocion = main_df.loc[main_df['Time_of_service'] > 15, 'Time_since_promotion'].max()
tiempo_spromocion_df = pd.DataFrame({'Tiempo_max_sin_promocion': [tiempo_spromocion]})

tiempo_spromocion_df
```

**La edad promedio en la que ingresaron las personas a la empresa, genera una nueva columna derivada mediante la formula Age - Time_of_service**

```python
#main_df.info()
edad_promedio_df = main_df.loc[:, ['Employee_ID', 'Age', 'Time_of_service']]
edad_promedio_df['Edad_Ingreso'] = edad_promedio_df['Age'] - edad_promedio_df['Time_of_service']
edad_promedio_df = edad_promedio_df[['Employee_ID', 'Age', 'Time_of_service', 'Edad_Ingreso']]

edad_promedio_df
```

**Un dataframe con las columnas Employee_ID, Relationship, Decision_skill_process que sólo contenga los registros de mujeres, solteras y con compensación tipo 2 o 3**

```python
filtrado_df = main_df.loc[(main_df['Gender'] == 'F') & (main_df['Relationship_Status'] == 'Single') & (main_df['Compensation_and_Benefits'] == 'type2'), ['Employee_ID', 'Relationship_Status', 'Decision_skill_possess']]

filtrado_df
```

**Una columna derivada con el valor "on time" si cuando tuvo su última promoción tenía menos de 30 años. En caso contrario colocar el valor de "out of time"**

```python
#main_df.info()

on_time_df = main_df.loc[:, ['Employee_ID', 'Age', 'Time_since_promotion']]
on_time_df['Age_ajustada'] = on_time_df['Age'] - on_time_df['Time_since_promotion']

on_time_df.loc[(on_time_df['Age_ajustada']< 30), 'Derivada'] = 'on time'
on_time_df.loc[(on_time_df['Age_ajustada']> 30), 'Derivada'] = 'out of time'
on_time_df = on_time_df[['Employee_ID', 'Derivada']]

on_time_df
```

**Guardar los dataframes resultantes**

```python
servicio_promedio_df.to_excel("A-Servicio Promedio.xlsx")
porcentaje_df.to_excel("B-Procentaje.xlsx")
tiempo_spromocion_df.to_excel("C-Tiempo Sin Promocion.xlsx")
edad_promedio_df.to_excel("D-Edad Promedio Ingreso.xlsx")
filtrado_df.to_excel("E-Mujeres Filtro.xlsx")
on_time_df.to_excel("F-Columna On Time.xlsx")
```

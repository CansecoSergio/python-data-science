7.	A través de la función `describe()` responde:
*  ¿Cuál es el tiempo máximo de servicio?
*  ¿Qué promedio de edad tienen los empleados encuestados?
*  ¿Cuál es el valor del 1er cuartil en la escala de pago?



```
stats_describe_df = employees_df.describe().T

print('Tiempo máximo de servicio: ', stats_describe_df.loc['Time_of_service', 'max'])
print('Promedio de edad empleados encuestados: ', round(stats_describe_df.loc['Age', 'mean'], 2))
print('Valor 1 er cuartil en la escala de pago: ', round(stats_describe_df.loc['Pay_Scale', '25%'], 2))
print('\n')
stats_describe_df
```

10.	Realiza una agrupación por ciudad, con el promedio de todas las columnas y almacena el resultado en un nuevo dataframe (`employees_by_hometown`).

```
numeric_columns = np.append(arr=employees_df.describe().columns.values, values='Hometown')
employees_by_hometown = employees_df[numeric_columns].groupby('Hometown').mean()
#print(employees_df.describe().T)
employees_by_hometown
```

11.	Visualiza el dataframe obtenido en una única gráfica (usando Pandas: `employees_by_hometown.plot()`). Como no es legible por la diferencia de escalas en las columnas, sepáralas en subplots. ¿Qué ciudad tiene el mayor índice de deserción? Observa el resto de sus variables

```
employees_by_hometown.plot(subplots=True, layout=(5,2), figsize=(11,10))
plt.show()
```

```
ciudad_mayor_desercion = employees_by_hometown.loc[employees_by_hometown['Attrition_rate'] == employees_by_hometown['Attrition_rate'].max()].index.values[0]
print('La ciudad con mayor índice de deserción es: ', ciudad_mayor_desercion)
```

13.	Elabora dos scatter plots (del dataframe original sin NaN) para determinar si hay correlación entre:
*   Edad y Tasa de deserción
*   Tiempo de servicio y Tasa de deserción

```
sns.scatterplot(x=employees_df['Age'],y=employees_df['Attrition_rate'])
plt.show()
print('\nCorrelación Edad y Tasa de deserción: ', round(employees_df['Age'].corr(employees_df['Attrition_rate']), 2))
```

```
sns.scatterplot(x=employees_df['Time_of_service'],y=employees_df['Attrition_rate'])
plt.show()
print('\nCorrelación Tiempo de servicio y Tasa de deserción: ', round(employees_df['Time_of_service'].corr(employees_df['Attrition_rate']), 2))
```

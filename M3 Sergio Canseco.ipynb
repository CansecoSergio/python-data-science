```python
import pandas as pd
import numpy as np
pd.set_option('display.max_rows', 10)
```

 **1.- Lectura de los archivos**

```python
df_test1 = pd.read_csv('/content/drive/MyDrive/Archivos Collab/M3/Test.csv')
df_test2 = pd.read_csv('/content/drive/MyDrive/Archivos Collab/M3/Test2.csv')

df_train1 = pd.read_csv('/content/drive/MyDrive/Archivos Collab/M3/Train.csv')
df_train2 = pd.read_csv('/content/drive/MyDrive/Archivos Collab/M3/Train2.csv')

print('DF TEST 1')
df_test1.info()
print('---')
print('DF TEST 2')
df_test2.info()
print('---')
print('DF TRAIN 1')
df_train1.info()
print('---')
print('DF TRAIN 2')
df_train2.info()
```

**2.- Uni칩n de los datos para los archivos con la informaci칩n de empleados.**

```python
employees_df = pd.concat([df_test1, df_train1], ignore_index=True)
employees_df
```

**3.- Uni칩n de los datos para los archivos con la informaci칩n de tickets.**

```python
tickets_df = pd.concat([df_test2, df_train2], ignore_index=True)
tickets_df.columns = tickets_df.columns.str.strip()
tickets_df.info()
```

```python
#count = employees_df[['Employee_ID']].groupby('Employee_ID').value_counts().reset_index(name='Count')
#max = count[['Count']].max()
#max
```

**4.- Join de ambos dataframes (empleados y tickets).**

```python
join_df = employees_df.merge(tickets_df, left_on='Employee_ID', right_on='Employee_ID', how='left')
#join_df[['Employee_ID', 'Computer_OS']]
join_df.info()
```

**5.- Asignaci칩n del valor "no registrado", para aquellos empleados que no hay registro de su sistema operativo.**

```python
join_df.loc[pd.isna(join_df['Computer_OS']), ['Computer_OS']] = 'no registrado'
join_df.loc[pd.isna(join_df['Mobile_OS']), ['Mobile_OS']] = 'no registrado'
join_df[['Employee_ID', 'Computer_OS', 'Mobile_OS']]
```

**A- El porcentaje de personas que usan "Windows/Mac/Linux/no registrado" como su sistema operativo en su computadora**

```python
df_a = join_df[['Employee_ID', 'Computer_OS']]
df_a = join_df.groupby(by='Computer_OS').agg({'Computer_OS': np.size}).reset_index(names='SO').rename(columns={'Computer_OS': 'Total'})
df_a['Porcentaje'] = (df_a['Total'] * 100) / len(join_df.index)
df_a = df_a[['SO', 'Porcentaje']]
df_a
```

**B- El porcentaje de personas que usan iOS / Android / No registrado como su sistema operativo en su dispositivo m칩vil**

```python
df_b = join_df[['Employee_ID', 'Mobile_OS']]
df_b = join_df.groupby(by='Mobile_OS').agg({'Mobile_OS': np.size}).reset_index(names='SO').rename(columns={'Mobile_OS': 'Total'})
df_b['Porcentaje'] = (df_b['Total'] * 100) / len(join_df.index)
df_b = df_b[['SO', 'Porcentaje']]
df_b
```

**C- La edad promedio de la gente que utiliza cada sistema operativo de escritorio**

```python
df_c = join_df[['Employee_ID', 'Age', 'Computer_OS']]
df_c = join_df.groupby(by='Computer_OS').agg({'Age': np.mean}).reset_index(names='SO').rename(columns={'Age': 'Edad_Promedio'})
df_c['Edad_Promedio'] = round(df_c['Edad_Promedio'], 2)
df_c
```

**D- El nivel educativo promedio de la gente que utiliza cada sistema operativo de escritorio**

```python
df_d = join_df[['Employee_ID', 'Education_Level', 'Computer_OS']]
df_d = join_df.groupby(by='Computer_OS').agg({'Education_Level': np.mean}).reset_index(names='SO').rename(columns={'Education_Level': 'Nivel_Educativo_Promedio'})
df_d['Nivel_Educativo_Promedio'] = round(df_d['Nivel_Educativo_Promedio'])
df_d
```

**E- El n칰mero total de tickets generados por la gente con cada sistema operativo de escritorio**

```python
df_e = join_df[['Employee_ID', 'Computer_tickets', 'Computer_OS']]
df_e = join_df.groupby(by='Computer_OS').agg({'Computer_tickets': np.sum}).reset_index(names='SO').rename(columns={'Computer_tickets': 'Total_TIckets'})
df_e
```

**6.- Almacenamiento archivo Excel**

```python
with pd.ExcelWriter('/content/drive/MyDrive/Archivos Collab/M3/resultados.xlsx') as writer:
    df_a.to_excel(writer, sheet_name='Ej A')
    df_b.to_excel(writer, sheet_name='Ej B')
    df_c.to_excel(writer, sheet_name='Ej C')
    df_d.to_excel(writer, sheet_name='Ej D')
    df_e.to_excel(writer, sheet_name='Ej E')
```
